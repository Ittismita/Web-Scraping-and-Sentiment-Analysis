#### Importing necessary libraries

```python
from bs4 import BeautifulSoup
import requests
import smtplib
import time
import datetime
```

```python
URL='https://en.wikipedia.org/wiki/List_of_largest_companies_in_India'

  
page=requests.get(URL)

soup1=BeautifulSoup(page.content,"html")
soup2=BeautifulSoup(soup1.prettify(),"html")
soup2

```

```python
#getting table at index 1
table_v1=soup2.find_all('table')[0]
print(table_v1)
```

```python
#getting all the of table at index 1
table_v2=table_v1.find_all('th')
print(table_v2)
```

```python
table_headers=[i.text.replace("\n","").replace(" ","") for i in table_v2]

table_headers
```

```python
import pandas as pd
```

```python
table_headers=['Rank', 'Name', 'Industry', 'Revenue(billionsUS$)', 'Headquarters']
```

```python
df=pd.DataFrame(columns=table_headers)
df
```

```python
row_data=table_v1.find_all('tr')
row_data
```

```python

for row in row_data[1:]:
    row=row.find_all('td')
    cell_data=[celldata.text.strip() for celldata in row]
    
    df.loc[len(df)]=cell_data
    
```

```python
df
```

```python
df.to_csv(r'IndianCompanies.csv',index=False)
```
